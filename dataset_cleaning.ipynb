```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

Beginning analysis of the Heart Disease Indicators 

```python
df = pd.read_csv("heartdisease_rates_trends.csv", dtype={'column_name': 'str'})
```

```python
df.info()
```

```python
df.head()
```

Dropping unneeded columns from first visual inspection

```python
df.drop(columns=['StratificationCategory1', 'StratificationCategory2', 
'StratificationCategory3',  'Data_Value_Footnote', 'GeographicLevel', 'DataSource', 'Data_Value_Type',
'Class', 'LocationID', 'Confidence_limit_Low', 'Confidence_limit_High', 'Data_Value_Footnote_Symbol'], inplace=True)

df.head()
```

```python
value_counts = df['Data_Value_Unit'].value_counts()
print("Count of each unique value:")
print(value_counts)
```

```python
# Rename the column from 'Data_Value_Unit' to 'DataValueUnit'
df = df.rename(columns={'Data_Value_Unit': 'DataValueUnit'})

# Retaining only instances of 100,000
df = df[df['DataValueUnit'] == 'per 100,000'].copy()
print("Unique values in DataValueUnit after renaming and filtering:")
print(df['DataValueUnit'].unique()) 
print("Number of rows after filtering:", len(df))
```

Check for missing values 

```python
total_rows = len(df)

# Calculating percentage of missing values for each column
missing_percentage = (df.isnull().sum() / total_rows) * 100
print(missing_percentage)
```

```python
# Dropping rows with missing values 
df = df.dropna(subset=['Data_Value'])

total_rows = len(df)
# Calculating percentage of missing values for each column
missing_percentage = (df.isnull().sum() / total_rows) * 100
print(missing_percentage)
```

Check of duplicates 

```python
duplicates = df[df.duplicated()]
num_duplicates = df.duplicated().sum()
print(f"Total number of duplicate rows: {num_duplicates}")
```

```python
# Dropping duplicate entries 
df = df.drop_duplicates()
```

```python
df.head()
```

Exploring spread of data across the years 

```python
value_counts = df['Year'].value_counts()
print(value_counts)
```

```python
# Keeping five years of spread 
# Subset of years to keep 
years_to_keep = [2016, 2018, 2019, 2015, 2017]
df = df[df['Year'].isin(years_to_keep)]


year_counts = df['Year'].value_counts()
# Percentage of each year as extra check 
year_percentage = (year_counts / len(df)) * 100
print("Percentage counts of unique values in 'Year' column:")
print(year_percentage)
```

```python
df.info()
```

Retaining top 5 states and making a column for state names. 

```python
df.head()
```

```python
value_counts = df['LocationAbbr'].value_counts()
print(value_counts)
```

```python
state_mapping = {'TX': 'Texas','GA': 'Georgia','VA': 'Virginia','NC': 'North Carolina','KY': 'Kentucky','MO': 'Missouri',
    'IL': 'Illinois','TN': 'Tennessee','OH': 'Ohio','FL': 'Florida','MS': 'Mississippi','IN': 'Indiana',
    'IA': 'Iowa','MI': 'Michigan','CA': 'California','OK': 'Oklahoma','KS': 'Kansas','MN': 'Minnesota',
    'AR': 'Arkansas','AL': 'Alabama','LA': 'Louisiana','NY': 'New York','PA': 'Pennsylvania',
    'NE': 'Nebraska','CO': 'Colorado','WI': 'Wisconsin','SC': 'South Carolina','WV': 'West Virginia',
    'WA': 'Washington','SD': 'South Dakota','MT': 'Montana','ID': 'Idaho','OR': 'Oregon','NM': 'New Mexico',
    'ND': 'North Dakota','MD': 'Maryland','NJ': 'New Jersey','UT': 'Utah','AZ': 'Arizona','MA': 'Massachusetts',
    'AK': 'Alaska', 'NV': 'Nevada','ME': 'Maine','VT': 'Vermont','CT': 'Connecticut','NH': 'New Hampshire',
    'RI': 'Rhode Island','HI': 'Hawaii','DE': 'Delaware','DC': 'District of Columbia'}

df['State'] = df['LocationAbbr'].map(state_mapping)

#Rename LocationAbbr to StateAbbr for clarity 
df.rename(columns={'LocationAbbr': 'StateAbbr'}, inplace=True)

#Rename LocationDesc to County for clarity 
df.rename(columns={'LocationDesc': 'County'}, inplace=True)

#Rename Data_Value to DataValue for clarity 
df.rename(columns={'Data_Value': 'DataValue'}, inplace=True)

#Rename Stratification1 to Age for clarity 
df.rename(columns={'Stratification1': 'Age'}, inplace=True)

#Rename Stratification2 to Race for clarity 
df.rename(columns={'Stratification2': 'Race'}, inplace=True)

#Rename Stratification3 to Sex for clarity 
df.rename(columns={'Stratification3': 'Sex'}, inplace=True)

# Visual check 
df.head()
```

```python
# Did not take Texas or Georgia as top state as it has significantly more entries than others 
# Selected states are Missouri, Tennessee, Virginia, North Carolina and Kentucky - retaining only these 
states_to_keep = ['MO', 'NC', 'TN', 'KY', 'VA']
df = df[df['StateAbbr'].isin(states_to_keep)]

# Check - spread of data across retained states 
percentage_counts = df['StateAbbr'].value_counts(normalize=True) * 100
print(percentage_counts)
```

Checking the types of columns and making alterations to reduce unneccessary load 

```python
df.info()
```

```python
# Converting to int for analysis
df['Year'] = df['Year'].astype(int)  

# Converting these to category to save space 
df['StateAbbr'] = df['StateAbbr'].astype('category')
df['State'] = df['State'].astype('category')
df['Topic'] = df['Topic'].astype('category')
df['DataValueUnit'] = df['DataValueUnit'].astype('category')
df['Sex'] = df['Sex'].astype('category')
df['Race'] = df['Race'].astype('category')
df['Age'] = df['Age'].astype('category')

df.info()
```

Check the value counts across the data 

```python
value_counts = df['Topic'].value_counts()
print(value_counts)
```

Checking for data outliers 

```python
# Get summary statistics for DataValue
summary_stats = df['DataValue'].describe()
print("Summary statistics for DataValue:")
print(summary_stats)
```

Saving cleaned data to a new file 

```python
df.to_csv('statesubset.csv', index=False)
```
